### Checking if that's any duplicated genes

```python
duplicates = 0
missing_values = 0

for file in files:

    file_df = pd.read_csv(file, sep='\t')

    if (
        file_df["Gene ID"].duplicated().sum() > 0
        or file_df["Gene Symbol"].duplicated().sum() > 0
    ):
        duplicates += 1

    if file_df.isna().sum().sum() > 0:
        missing_values += 1

print("Duplicated 'Gene ID' or 'Gene Symbol':", duplicates)
print("Files with missing values:", missing_values)
```

### Filtering Non-Coding Genes

The original RNA-seq files contain multiple transcript types, including non-coding RNAs, pseudogenes, and other non-protein-coding features.

Since the main objective of this project is to investigate biologically interpretable gene expression patterns associated with dengue severity, only genes classified as `"mRNA"` were retained.

This filtering step reduces noise and focuses the analysis on protein-coding genes, which are typically more suitable for downstream differential expression and pathway analysis.

```python
for file in files:
    file_df = pd.read_csv(file, sep="\t")
    file_df = file_df[file_df["Type"] == "mRNA"]

    output_path = Path("../data/interim") / file.name
    file_df.to_csv(output_path, sep="\t", index=False)
```

```python
files = sorted(
    Path("../data/raw").glob("*.xls")
)

total_removed = 0

for file in files:

    file_df = pd.read_csv(file, sep="\t")

    original_rows = len(file_df)

    # Keep only protein-coding genes
    file_df = file_df[file_df["Type"] == "mRNA"]

    filtered_rows = len(file_df)
    removed_rows = original_rows - filtered_rows
    total_removed += removed_rows

    output_path = Path("../data/interim") / file.name

    file_df.to_csv(
        output_path,
        sep="\t",
        index=False
    )

    print(
        f"{file.name}: "
        f"removed {removed_rows} non-coding entries"
    )

print("\nTotal removed entries:", total_removed)
```

## Build the TPM Matrix

```python
from pathlib import Path
import pandas as pd

# Get all files
files = sorted(Path("../data/raw").glob("*.xls"))

# Load first file as base
base_df = pd.read_csv(files[0], sep="\t")

# Keep annotation columns
expression_matrix = base_df[["Gene ID", "Gene Symbol"]].copy()

# Add TPM columns
for file in files:
    df = pd.read_csv(file, sep="\t")

    # Find TPM column automatically
    tpm_col = [col for col in df.columns if "TPM" in col][0]

    # Extract sample name
    sample_name = tpm_col.replace(" TPM", "")

    # Add TPM values to matrix
    expression_matrix[sample_name] = df[tpm_col]

    print(f"Added: {sample_name}")

expression_matrix = expression_matrix.set_index("Gene ID")

# Save matrix
expression_matrix.to_csv(
    "../data/processed/expression_matrix_tpm.csv",
    index=False
)

print("\nTPM expression matrix saved successfully!")
```
