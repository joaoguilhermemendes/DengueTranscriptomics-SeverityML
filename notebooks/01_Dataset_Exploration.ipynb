# Dataset Summary — GSE279208

**Dataset: GSE279208**

**Title**:
Unraveling potential gene biomarkers for dengue infection through RNA sequencing

**Organism**:
Homo sapiens

**Technology**:
Bulk RNA sequencing (RNA-seq)

**Platform**:
DNBSEQ-G400 (MGI)

**Biological Material**:
Peripheral Blood Mononuclear Cells (PBMCs)

**Samples**:
26 total samples

**Groups**:
* Healthy controls (10 samples)
* Dengue with warning signs (13 samples)
* Severe dengue (3 samples)

**Objective**:
Identify host transcriptomic biomarkers associated with dengue infection severity using RNA-seq data.

**Study Design**:
PBMCs were isolated from whole blood samples, followed by RNA extraction, mRNA enrichment, library preparation, and high-throughput sequencing.

**Main Analyses in Original Paper**:
* Differentially expressed genes (DEGs)
* GO enrichment analysis
* KEGG pathway analysis

**Potential Biomarkers Highlighted**:
IFI27, KCTD14, SDC1, UCHL1, CCL2, TNFSF15, SERPINB2, ACOD1, among others.

**Paper**:
Suppiah, J., Md Sani, S. S., Hassan, S. S., Fasohah Nadzar, N. I., Ibrahim, I., Thayan, R., & Zain, R. M. (2024). Unraveling potential gene biomarkers for dengue infection through RNA sequencing. Virus Genes, 61(1), 26. https://doi.org/10.1007/s11262-024-02114-2
